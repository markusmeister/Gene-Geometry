## Plotting routines

````python
def plot(X, Y=None, xlabel=None, ylabel=None, legend=[], loc=None, title=None,
         xlim=None, ylim=None, xscale='linear', yscale='linear',
         xticks=None, yticks=None, xhide=False, yhide=False, yrot=False, 
         xzero=False, yzero=False, yflip=False, 
         fmts=['r-','g-','b-','y-','c-','m-'], linewidth=2, markersize=5, fillstyle='full',
         markeredgewidth=1, rasterized=False,
         grid=False, equal=False, figsize=(4,2.5), axes=None):
    """
    Plot data points.
    X: an array or list of arrays
    Y: an array or list of arrays
    If Y exists then those values are plotted vs the X values
    If Y doesn't exist the X values are plotted
    xlabel, ylabel: axis labels
    legend: list of labels for each Y series
    loc: location of the legend, like 'upper right'
    title: duh
    xlim, ylim: [low,high] list of limits for the 2 axes 
    xscale, yscale: 'linear' or 'log'
    xticks, yticks: list of locations for tick marks, or None for auto ticks
    yhide: hide the y axis?
    yrot: rotate the yaxis label to horizontal?
    yzero: zero line for the y-axis?
    fmts: a list of format strings to be applied to successive Y-series
    linewidth, markersize, fillstyle, markeredgewidth: see docs
    grid: draw a grid?
    equal: use equal aspect ratio, i.e. same scale per unit on x and y axis?
    figsize: (h,v) in inches
    rasterized: rasterize the data portion of the graph?
    axes: pre-existing axes where to draw the plot
    Returns: axes for the plot
    """
    
    if not axes: # start a new figure
        fig = plt.figure(figsize=figsize)
        axes = plt.gca()
    
    def has_one_axis(X): # Return True if X (ndarray or list) has 1 axis
        return (hasattr(X, "ndim") and X.ndim == 1 or isinstance(X, list)
                and not hasattr(X[0], "__len__"))

    if has_one_axis(X):
        X = [X]
    if Y is None:
        X, Y = [[]] * len(X), X
    elif has_one_axis(Y):
        Y = [Y]
    if len(X) != len(Y):
        X = X * len(Y)
    # axes.cla() # clears these axes
    for x, y, fmt in zip(X, Y, fmts):
        if len(x):
            axes.plot(x, y, fmt, linewidth=linewidth, markersize=markersize,
            	fillstyle=fillstyle,markeredgewidth=markeredgewidth,
            	rasterized=rasterized)
        else:
            axes.plot(y, fmt, linewidth=linewidth, markersize=markersize,
            	fillstyle=fillstyle,markeredgewidth=markeredgewidth,
            	rasterized=rasterized)
    set_axes(axes, xlabel, ylabel, legend, loc, xlim, ylim, xscale, yscale, 
             xticks, yticks, xhide, yhide, yrot, xzero, yzero, yflip, grid, equal)
    if title:
        plt.title(title)
    plt.tight_layout()

    return axes # useful if we started a new figure

def hist(X, bins=50, range=None, density=None, weights=None, histtype='stepfilled', stacked=False, 
         align='mid', color=['r','g','b','m'], alpha=0.5, linewidth=1, 
         xlabel=None, ylabel=None, legend=[], loc=None, title=None,
         xlim=None, ylim=None, xscale='linear', yscale='linear',
         xticks=None, yticks=None, xhide=False, yhide=False, yrot=False, 
         xzero=False, yzero=False, yflip=False, 
         grid=False, figsize=(4,2.5), axes=None, orientation='vertical'):
    """
    Histogram data.
    X: an array or list of arrays to be histogrammed
    bins: the number of bins, or a list of edges
    range: [low,high] the range over which the bins are spread
    density: normalize to unit area?
    weights: array same shape as X; each value in X only contributes that weight.
    histtype: 'bar','barstacked','step','stepfilled'
    stacked: plot multiple series on top of each other?
    align: 'left','mid',right', center the bar on left edge, center, or right edge of bin
    color: list of colors for the data series
    alpha: transparency, useful if histtype='stepfilled'
    xlabel, ylabel: axis labels
    legend: list of labels for each Y series
    loc: location of the legend, like 'upper right'
    title: duh
    xlim, ylim: [low,high] list of limits for the 2 axes 
    xscale, yscale: 'linear' or 'log'
    xticks, yticks: list of locations for tick marks, or None for auto ticks
    yhide: hide the y axis?
    yrot: rotate the yaxis label to horizontal?
    yzero: zero line for the y-axis?
    grid: draw a grid?
    figsize: (h,v) in inches
    axes: pre-existing axes where to draw the plot
    Returns: axes, n, bins, patches 
    """
    
    if not axes: # start a new figure
        fig = plt.figure(figsize=figsize)
        axes = plt.gca()
    
    def has_one_axis(X): # Return True if X (ndarray or list) has 1 axis
        return (hasattr(X, "ndim") and X.ndim == 1 or isinstance(X, list)
                and not hasattr(X[0], "__len__"))

    if has_one_axis(X):
        ndata=1
    else:
        ndata=len(X) # assuming it is an ndarray or a list of lists
    
    if len(color)<ndata: # if there aren't enough dolors provided
        col=None # use the default color sequence
    else:
        col=color[:ndata]
    
    if histtype=='step' or histtype=='stepfilled': # in these cases the legend sequence needs to be inverted for some reason
        if isinstance(legend, list):
        	legend = legend[-1::-1]
    
    n, bins, patches = axes.hist(X, bins=bins, range=range, density=density, weights=weights, 
    		histtype=histtype, stacked=stacked, align=align, 
            color=col, alpha=alpha, linewidth=linewidth,
            label=legend, orientation=orientation)

    set_axes(axes, xlabel, ylabel, legend, loc, xlim, ylim, xscale, yscale, 
    	xticks, yticks, xhide, yhide, yrot, xzero, yzero, yflip, grid, equal=False)

    if title:
        plt.title(title)
    plt.tight_layout()

    return axes, n, bins, patches # useful if we started a new figure

def set_axes(axes, xlabel, ylabel, legend, loc, xlim, ylim, xscale, yscale, 
    	xticks, yticks, xhide, yhide, yrot, xzero, yzero, yflip, grid, equal):
    """Set the axes for matplotlib."""
    axes.set_xscale(xscale)
    axes.set_yscale(yscale)
    if xlim:
        axes.set_xlim(xlim)
    else:
        axes.set_xlim(auto=True)
    if ylim:
        axes.set_ylim(ylim)
    else:
        axes.set_ylim(auto=True)
    if grid:
        axes.grid()
    if equal:
        axes.set_aspect(aspect='equal')
    if ylabel:
        if yrot:
            axes.set_ylabel(ylabel, fontsize=12, rotation=0, labelpad=15)
        else:
            axes.set_ylabel(ylabel, fontsize=12)
    if xlabel:
        axes.set_xlabel(xlabel, fontsize=12)
    axes.get_yaxis().set_visible(not yhide)
    axes.get_xaxis().set_visible(not xhide)
    if xzero:
        axes.axvline(color='black', linewidth=0.5)
    if yzero:
        axes.axhline(color='black', linewidth=0.5)
    if yflip:
        axes.invert_yaxis()
    axes.tick_params(axis = 'both', which = 'major', labelsize = 10)
    axes.tick_params(axis = 'both', which = 'minor', labelsize = 9)
    if xticks:
        axes.set_xticks(xticks,minor=False); # no minor ticks
    if yticks:
        axes.set_yticks(yticks,minor=False); # no minor ticks
    if legend:
        axes.legend(legend, loc=loc)
    plt.draw()
````

````python
# make all permutations of corners of a cube
Mac_Perm=[]
i=0
for j in range(1,8):
    for k in range(j+1,8):
        for l in range(k+1,8):
            rl=[x for x in list(range(1,8)) if x!=j and x!=k and x!=l] # 4 left
            for m in rl:
                rm=[x for x in rl if x!=m] # 3 left
                for n in rm:
                    rn=[x for x in rm if x!=n] # 2 left
                    for o in rn:
                        p=[x for x in rn if x!=o][0] # 1 left
                        Mac_Perm+=[(i,j,k,l,m,n,o,p)]
con=[0,1,2,4,3,6,5,7] # convert the corners from the permutation list to conventional labeling
Mac_Perm=[[x[c] for c in con] for x in Mac_Perm]
````

````python
# Compute the fraction of edges shared with the "correct" cube for each permutation
#edges on the cube in standard coordinates
E=[[0,1],[0,2],[0,4],[1,3],[1,5],[2,3],[2,6],[4,5],[4,6],[3,7],[5,7],[6,7]]

def adj(p):
    '''
    returns adjacency matrix for cube permuted by p
    '''
    A=np.zeros((8,8))
    for [i,j] in E:
        A[p[i],p[j]]=A[p[j],p[i]]=1
    return A
    
B=adj((0,1,2,3,4,5,6,7)) # correct cube
# for each permutation in per compute fraction of 1's correct in the adjacency matrix
Mac_Perm_Fred=np.array([np.sum(adj(p)*B)/24 for p in Mac_Perm]) # fraction of edges in each permutation
Mac_Perm_OK=np.argmax(Mac_Perm_Fred) # the correct permutation
````

# Cartesian model

## How to fit a Cartesian shape to the data

Assuming a 3D model, e.g. $2 \times 2 \times 2$. 

Data points:

$$
\vec x_{i,j,k}, \quad i \in \{1,\dots,L\},j \in \{1,\dots,M\},k \in \{1,\dots,N\}
$$

where $i,j,k$ index the 3 phenotypic dimensions, for example "fovea/periphery", "on/off", "transient/sustained".

Fit:

$$
\vec y_{i,j,k} = \vec a + \vec b_i + \vec c_j + \vec d_k
$$

The model parameters are the vectors $\vec a, \{\vec b_i\}, \{\vec c_j\}, \{\vec d_k\}$. For uniqueness we require that

$$\sum_i \vec b_i = \sum_j \vec c_j = \sum_k \vec d_k = 0$$

Squared residual:

$$
\chi^2 = \sum_{i,j,k} \left( \vec x_{i,j,k} - \vec y_{i,j,k} \right)^2
$$

The residual is minimized by this solution:

\begin{align*}
\vec a &= \left< \vec x_{i,j,k} \right>_{i,j,k} = \frac{1}{LMN} \sum_{i,j,k} \vec x_{i,j,k} \\
\vec b_i &= \left< \vec x_{i,j,k} - \vec a \right>_{j,k} = \frac{1}{MN} \sum_{j,k} \left( \vec x_{i,j,k} - \vec a \right) \\
\vec c_j &= \left< \vec x_{i,j,k} - \vec a \right>_{i,k} = \frac{1}{LN} \sum_{i,k} \left( \vec x_{i,j,k} - \vec a \right) \\
\vec d_k &= \left< \vec x_{i,j,k} - \vec a \right>_{i,j} = \frac{1}{LM} \sum_{i,j} \left( \vec x_{i,j,k} - \vec a \right)
\end{align*}

The total variance of the data is

$$
V_{\rm{tot}} = \frac{1}{LMN} \sum_{i,j,k} \left( \vec x_{i,j,k} - \vec a \right)^2
$$

and the variance explained by the model is

$$
V_{\rm{exp}} = \left< \left( \vec y_{i,j,k} - \vec a \right)^2 \right>_{i,j,k} = \frac{1}{LMN} \sum_{i,j,k} \left(\vec b_i + \vec c_j + \vec d_k\right)^2
$$

For some purposes we will want to consider only part of the variance, for example the variance within each species. Assume that the last dimension (spanned by the vectors $\vec d_k$) is the species dimension. Then the mean within each species is

$$
\vec a_k = \left< \vec x_{i,j,k} \right>_{i,j} = \frac{1}{LM} \sum_{i,j} \vec x_{i,j,k}
$$

and the intraspecies variance of the data is

$$
V^{\rm{(s)}} = \frac{1}{LMN} \sum_{i,j,k} \left( \vec x_{i,j,k} - \vec a_k \right)^2
$$

and the intraspecies variance explained by the model is

$$
V^{\rm{(s)}}_{\rm{exp}} = \left< \left( \vec y_{i,j,k} - \vec a_k \right)^2 \right>_{i,j,k} = \frac{1}{LMN} \sum_{i,j,k} \left(\vec b_i + \vec c_j\right)^2
$$








The solution generalizes in obvious ways to lower or higher dimensions. For the model structures used in the analysis so far:

$2 \times 2 \times 2$: This is a 3D model with $L=2,M=2,N=2$

$4 \times 2$: This is a 2D model with $L=4,M=2$. There are no $\vec d_k$ vectors.

$4 \times 3$: This is a 2D model with $L=4,M=3$.



## 2x2x2 Model

````python
def f_2x2x2(X,type_list,n_dims):
    '''
    Performs 2x2x2 Cartesian model fit
    X: data matrix (n_features x n_types)
    type_list: list of 8 types in order [nT1,nS1,fT1,fS1,nT2,nS2,fT2,fS2]
    n_dims: number of features of X to use
    returns: fraction of variance explained by the model, both total and within species
    '''
    Z=X[type_list,:n_dims] 
    a=np.sum(Z,axis=0)/8 # mean
    Z-=a.reshape(-1,len(a)) # subtract mean
    V=np.sum(Z**2) # total variance
    Z2=Z.copy()
    Z2[:4]-=(np.sum(Z[:4],axis=0)/4).reshape(-1,len(a)) # subtract the mean within the last dimension
    Z2[4:]-=(np.sum(Z[4:],axis=0)/4).reshape(-1,len(a))
    V2=np.sum(Z2**2) # variance excluding the last dimension
    B=np.array([(Z[i]+Z[i+2]+Z[i+4]+Z[i+6])/4 for i in [0,1]]) # the 2 vectors for the 1st dimension
    C=np.array([(Z[i]+Z[i+1]+Z[i+4]+Z[i+5])/4 for i in [0,2]]) # the 2 vectors for the 2nd dimension
    D=np.array([(Z[i]+Z[i+1]+Z[i+2]+Z[i+3])/4 for i in [0,4]]) # the 2 vectors for the 3rd dimension
    Ve=np.sum([np.sum((b+c+d)**2) for d in D for c in C for b in B]) # explained variance
    Ve2=np.sum([np.sum((b+c)**2) for d in D for c in C for b in B]) # explained variance excluding the last dimension    
    return Ve/V,Ve2/V2
````

## 4x2 Model

````python
def f_4x2(X,type_list,n_dims):
    '''
    Performs 4x2 Cartesian model fit
    X: data matrix (n_features x n_types)
    type_list: list of 8 types in order [nT1,nS1,fT1,fS1,nT2,nS2,fT2,fS2]
    n_dims: number of features of X to use
    returns: fraction of variance explained by the model, both total and within species
    '''
    Z=X[type_list,:n_dims] 
    a=np.sum(Z,axis=0)/8 # mean
    Z-=a.reshape(-1,len(a)) # subtract mean
    V=np.sum(Z**2) # total variance
    Z2=Z.copy()
    Z2[:4]-=(np.sum(Z[:4],axis=0)/4).reshape(-1,len(a)) # subtract the mean within the last dimension
    Z2[4:]-=(np.sum(Z[4:],axis=0)/4).reshape(-1,len(a))
    V2=np.sum(Z2**2) # variance excluding the last dimension
    B=np.array([(Z[i]+Z[i+4])/2 for i in range(4)]) # the 4 vectors in the 1st dimension
    C=np.array([np.sum(Z[i:i+4],axis=0)/4 for i in [0,4]]) # 2 vectors in the 2nd dimension
    Ve=np.sum([np.sum((b+c)**2) for c in C for b in B])
    Ve2=np.sum([np.sum((b)**2) for c in C for b in B]) # explained variance excluding the last dimension    
    return Ve/V,Ve2/V2
````

## 4x3 Model

````python
def f_4x3(X,type_list,n_dims):
    '''
    Performs 4x3 Cartesian model fit
    X: data matrix (n_features x n_types)
    type_list: list of 12 types in order [nT1,nS1,fT1,fS1,nT2,nS2,fT2,fS2,nT3,nS3,fT3,fS3]
    n_dims: number of features of X to use
    returns: fraction of variance explained by the model, both total and within species
    '''
    Z=X[type_list,:n_dims] 
    a=np.sum(Z,axis=0)/12 # mean
    Z=Z-a.reshape(-1,len(a)) # subtract the mean
    V=np.sum(Z**2) # total variance
    Z2=Z.copy()
    Z2[:4]-=(np.sum(Z[:4],axis=0)/4).reshape(-1,len(a)) # subtract the mean within the last dimension
    Z2[4:8]-=(np.sum(Z[4:8],axis=0)/4).reshape(-1,len(a))
    Z2[8:]-=(np.sum(Z[8:],axis=0)/4).reshape(-1,len(a))
    V2=np.sum(Z2**2) # variance excluding the last dimension
    B=np.array([(Z[i]+Z[i+4]+Z[i+8])/3 for i in range(4)]) # the 4 vectors for the 1st dimension
    C=np.array([np.sum(Z[j*4:j*4+4],axis=0)/4 for j in range(3)]) # the 3 vectors for the 2nd dimension
    Ve=np.sum([np.sum((b+c)**2) for c in C for b in B])
    Ve2=np.sum([np.sum((b)**2) for c in C for b in B]) # explained variance excluding the last dimension    
    return Ve/V,Ve2/V2
````

````python
# Compute the SNR with respect to the known 2x2x2 Cartesian product
SNR=[]
(i,j,k,l,m,n,o,p)=(0,1,2,3,4,5,6,7)
for Z,S in zip(Mac_Mean,Mac_SEM):
    b=(Z[i]-Z[j]+Z[k]-Z[l]+Z[m]-Z[n]+Z[o]-Z[p])/8 # weight corresponding to the first dimension
    c=(Z[i]+Z[j]-Z[k]-Z[l]+Z[m]+Z[n]-Z[o]-Z[p])/8 # weight corresponding to the second dimension
    d=(Z[i]+Z[j]+Z[k]+Z[l]-Z[m]-Z[n]-Z[o]-Z[p])/8 # weight corresponding to the third dimension
    Sig=b**2+c**2+d**2 # explained variance for 2x2x2 geometry
    Noi=np.sum(S**2)/8 # variance within clusters
    SNR+=[Sig/max(Noi,1e-10)]
Mac_SNR_222=np.array(SNR)
````

For the present purpose, the macaque cells stay as a constant reference, and we are evaluating different mouse partners for them by how well the shape among mouse types matches the shape among the macaque types.

## Restrict the score to the intra-species explained variance

````python
def plot_Mou_Mac_4(mac_types,model,N_genes,N_dims):
    '''
    Like plot_Mou_Mac() but uses only intra-species explained variance
    Fits mouse types (all combinations) to the specified macaque types
    mac_types = list of macaque types
    model = choice of model for fitting
    N_genes = list of values of n_genes
    N_dims = list of values of n_dims
    returns: list of explained variances, averaged over conditions 
    '''
    fve=np.zeros(len(Mou_Com)) # fraction variance explained
    for n_genes in N_genes:
        Ge=np.argsort(Mix_SNR)[-n_genes:] # rows of the gene list that we will use
        D=Mix_Mean[Ge].T # rows are types
        X=PCA().fit_transform(D) # all components
        for n_dims in N_dims:
            res=np.array([model(X,mou_types+mac_types,n_dims) for mou_types in Mou_Com]) # result of fit for every combination
            fve+=res[:,1] # the explained variance within species
    fve/=(len(N_genes)*len(N_dims)) # average explained variance over fit parameters
    fve_4=fve[np.array(Mou_Nalpha)==4]
    fve_3=fve[np.array(Mou_Nalpha)==3]
    fig,ax=plt.subplots(2,1,figsize=(4.5,2.5),height_ratios=[3,0.5],sharex=True,
        gridspec_kw={'hspace':0.0},
    );
    bins=np.linspace(0.2,0.7,101)
    hist([fve_4,fve_3,fve],color=['red','g','lightgray'],alpha=1,bins=bins,axes=ax[0]); # histogram the explained variance values
    ax[0].get_xaxis().set_visible(False)
    ax[0].set_title(model.__name__+f', n_genes in {N_genes}, n_dims in {N_dims}',fontsize=8);
    ax[0].set_ylabel('Combinations',fontsize=9);
    # mark the combinations containing alphas with a blue line, zetas with a green line
    segs=np.array([[[r,0],[r,0+n]] for r,n in zip(fve,Mou_Nalpha) if n>0]) # alphas
    lines=LineCollection(segs,color='blue',linewidth=0.5)
    ax[1].add_collection(lines);
#     mi=fve.min();ma=fve.max();ax[1].set_xlim(mi-0.05*(ma-mi),ma+0.05*(ma-mi))
    ax[1].set_xlim(0.2,0.68)
    ax[1].set_ylim(0,5)
    ax[1].get_yaxis().set_visible(False)
    ax[1].set_xlabel('Explained variance',fontsize=9);
    plt.tight_layout()
    return fve
````

## 3D geometry of the 4x2 Alpha-Fovea product

We can display the 4 fovea types and 4 mouse types in the 3D projection of the 4x2 model fit.

- Compute the 4x2 model fit
- Get the 4 vectors spanning the first dimension of the Cartesian product, they sum to zero
- project into the 3D space of those 4 vectors
- project the data into the same space

````python
# fovea and mouse Alpha
n_dims=12 # number of PC dimensions to use
n_genes=200
Ge=np.argsort(Mix_SNR)[-n_genes:] # rows of the gene list that we will use
D=Mix_Mean[Ge].T # rows are types
X=PCA().fit_transform(D) # all components
type_list=Mac_Fov+Mou_Alpha
Z=X[type_list,:n_dims] 
````

````python
# fit 4x2
a=np.sum(Z,axis=0)/8 # mean
Z-=a.reshape(-1,len(a)) # subtract mean
B=np.array([(Z[i]+Z[i+4])/2 for i in range(4)]) # the 4 vectors in the 1st dimension
C=np.array([np.sum(Z[i:i+4],axis=0)/4 for i in [0,4]]) # 2 vectors in the 2nd dimension
````

````python
# compute explained intra-species variance
Z2=Z.copy()
Z2[:4]-=(np.sum(Z[:4],axis=0)/4).reshape(-1,len(a)) # subtract the mean within the last dimension
Z2[4:]-=(np.sum(Z[4:],axis=0)/4).reshape(-1,len(a))
V2=np.sum(Z2**2) # variance excluding the last dimension
Ve2=np.sum([np.sum((b)**2) for c in C for b in B]) # explained variance excluding the last dimension    
Ve2/V2
````

````python
# corners of the fit shape
E=np.array([b+c for c in C for b in B])
````

````python
pca=PCA(n_components=3)
E2=pca.fit_transform(B) # transform into the 3 dimensions spanned by the 4 shape vectors
E2=np.concatenate((E2,E2),axis=0) # duplicate the 4 vectors
# transform the 4 fovea points and the 4 periphery points the same way
# subtract their mean first
Z0=pca.transform(Z[:4]-(np.sum(Z[:4],axis=0).reshape(-1,len(a))/4)) 
Z1=pca.transform(Z[4:]-(np.sum(Z[4:],axis=0).reshape(-1,len(a))/4))
Z2=np.concatenate((Z0,Z1),axis=0) # concatenate to 8 points for easy plotting
````

## Which genes contribute the most to the Alpha-Macaque match? (in progress)
One can ask which genes carry the most signal about the proposed alpha/macaque geometry. Obviously this is a post-hoc analysis after identifying the Alpha cells as partners.

````python
SNR=[]
(i,j,k,l,m,n,o,p)=Mou_Alpha+Mac_Fov # mouse alpha and macaque fovea
for Z,S in zip(Mix_Mean,Mix_SEM):
    b=(Z[i]-Z[j]+Z[k]-Z[l]+Z[m]-Z[n]+Z[o]-Z[p])/8 # weight corresponding to the first dimension
    c=(Z[i]+Z[j]-Z[k]-Z[l]+Z[m]+Z[n]-Z[o]-Z[p])/8 # weight corresponding to the second dimension
    d=(Z[i]+Z[j]+Z[k]+Z[l]-Z[m]-Z[n]-Z[o]-Z[p])/8 # weight corresponding to the third dimension
    Sig=b**2+c**2+d**2 # explained variance for 2x2x2 geometry
    Noi=np.sum(S**2)/8 # variance within clusters
    SNR+=[Sig/max(Noi,1e-10)]
Mix_SNR_222=np.array(SNR)
````
